# 1.4 How to Display Image Dimensions?
RGB image has three color channels, a pixel value has three numbers, each ranging from 0 to 255 (both inclusive). 


In order to display image dimensions numpy will be used. For more information about numpy: https://numpy.org/doc/stable/

```python
import numpy as np
print(image.shape)

#In order to access a dimension:
print('Height : {} pixels.'.format(int(image.shape[0])))
print('Width : {} pixels.'.format(int(image.shape[1])))
print('Depth : {} pixels.'.format(int(image.shape[2])))
```

# 2. Grayscaling Images

```python
image = cv2.imread('images/castara.jpeg')
```

```python
def show(image=None,title=''):
        size = 8
        w,h = image.shape[0],image.shape[1]
        aspect_ratio = w/h # a proportional relationship between an image's width and height.
        plt.figure(figsize=(size*aspect_ratio,size))
        plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
        plt.title("Image")
        plt.show
```

```python
show(image,'RGB Image')
```

```python
gray_scale = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) #cvtColor is used to convert the image to grayscale
show(gray_scale,'Grayscaled')
```

```python
# for the grayscale image, image dimensions can be expressed as  1WH
gray_scale.shape
```

# 3.2 HSV Color Space
